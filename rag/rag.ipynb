```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('C:\\Users\\kousi\\Documents\\hackathon\\documents\\Mental_1.pdf')
```

```python
data = loader.load()
data
```

```python
content = ''
for page in data:
    content += page.page_content
print(content)
```

```python
import re
text = content
text = re.sub(r'<.*?>', ' ', text)
text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ', text)
text = re.sub(r'\\s+', ' ', text)
text = re.sub(r'\n+', ' ', text)
text = re.sub(r'\.+', ' ', text)
text = re.sub(r'\t+', ' ', text)
text = re.sub(r'\-+', ' ', text)
text = re.sub(r'\S*@\S*\s?', ' ', text)
text = re.sub(r'@\w+', ' ', text)   # mentions
text = re.sub(r'#\w+', ' ', text)   # hashtags (optional: keep word)
text = re.sub(r'\d+', ' ', text)           # remove
text = re.sub(r'\d+', '<NUM>', text)       # replace with token
```

```python
from langchain_core.output_parsers import StrOutputParser

system_prompt = """
You are a helpful assistant that can answer questions about Mental Health issues.
"""

user_question = "I am facing mental health issues. Based on the given text, can you suggest some coping strategies and resources that might help me manage my condition better?"

human_prompt = f"""
{user_question}
"""
from langchain_core.prompts import ChatPromptTemplate , SystemMessagePromptTemplate , HumanMessagePromptTemplate

system_message_prompt = SystemMessagePromptTemplate.from_template(system_prompt)

human_message_prompt = HumanMessagePromptTemplate.from_template(human_prompt)

prompt = ChatPromptTemplate.from_messages([system_message_prompt , human_message_prompt])

chain = prompt | llm | StrOutputParser()

chain.invoke({"user_question":user_question, "content":content})
```

```python
from langchain_unstructured import UnstructuredLoader

page_url = "https://www.who.int/news-room/fact-sheets/detail/mental-health-strengthening-our-response"
page_url_1 = "https://www.thebridgelab.org/mental-wellness-resources/mh/what-is-stress"
page_url_2 = "https://www.thebridgelab.org/mental-wellness-resources/mh/distress"
page_url_3 = "https://www.thebridgelab.org/mental-wellness-resources/mh/burn-out"
page_url_1 = "https://www.thebridgelab.org/mental-wellness-resources/mh/preventing-burn-out"


all_urls = [page_url, page_url_1, page_url_2, page_url_3, page_url_1]

docs = []
for i in all_urls:
    loader = UnstructuredLoader(web_url=i)
    async for doc in loader.alazy_load():
        docs.append(doc)
```

Here i am following a approach to build a vector store to store my chunks there

```python
from langchain_community.document_loaders import UnstructuredURLLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
# chunks = splitter.split_documents(docs)
chunks = splitter.split_documents(data)
```

```python
!pip install chromadb
```

```python
from langchain_community.vectorstores import Chroma
embeddings = OllamaEmbeddings(
        model="nomic-embed-text",   
        base_url="http://localhost:11434" 
    )

vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="C:\\Users\\kousi\\Documents\\hackathon\\rag\\vectorstore"
)
vectorstore.persist() 
```

```python
PERSIST_DIRECTORY = "C:\\Users\\kousi\\Documents\\hackathon\\rag\\vectorstore"
db = Chroma(
    persist_directory=PERSIST_DIRECTORY,
    embedding_function=embeddings
)
query="what to do if i am facing mental health issues?"
docs = db.similarity_search(query, k=4)
```

```python
answer = ''
for i in range(len(docs)):
    answer += docs[i].page_content
```

```python
def rewrite(answer):
    prompt = f"Please rewrite the following answer to be more empathetic and supportive:\n\n{answer}\n\nRewritten Answer:"
    rewritten_answer = llm.invoke(prompt)
    return rewritten_answer

rewrite(answer)
```
