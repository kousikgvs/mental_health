```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('C:\\Users\\kousi\\Documents\\hackathon\\documents\\Mental_1.pdf')
```

```python
data = loader.load()
data
```

```python
content = ''
for page in data:
    content += page.page_content
print(content)
```

```python
import re
text = content
text = re.sub(r'<.*?>', ' ', text)
text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ', text)
text = re.sub(r'\\s+', ' ', text)
text = re.sub(r'\n+', ' ', text)
text = re.sub(r'\.+', ' ', text)
text = re.sub(r'\t+', ' ', text)
text = re.sub(r'\-+', ' ', text)
text = re.sub(r'\S*@\S*\s?', ' ', text)
text = re.sub(r'@\w+', ' ', text)   # mentions
text = re.sub(r'#\w+', ' ', text)   # hashtags (optional: keep word)
text = re.sub(r'\d+', ' ', text)           # remove
text = re.sub(r'\d+', '<NUM>', text)       # replace with token
```
